```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
import scipy.stats as st
import scipy.ndimage as nd
import scipy.signal as sg
from matplotlib.ticker import EngFormatter
from scipy.optimize import curve_fit


def init(filename):
    file=sio.loadmat(filename) #On charge le fichier
    Nbobjets = int(max(file['objs_link'][3]))
    print("On a tracqué "+str(Nbobjets)+" particules")
    return file,Nbobjets
file, Nbobjets=init("results/dernier.mat")
frequence=400
longueurminimale=1000
binning=1
agrandissement=20
diametre=1 #Taille des billes en µm
temp=293 #Temperature en K
kb=1.38e-11 #Constante de Boltzmann en unités adaptées

def take(i):
    """La fonction qui tracque la particule i"""
    PixelSize=binning*(6450/agrandissement)*1e-3 # en um
    x1=file['objs_link'][0,:][np.where(file['objs_link'][5,:]==i)]*PixelSize
    y1 = file['objs_link'][1,:][np.where(file['objs_link'][5,:]==i)]*PixelSize
    dureetrajectoire=x1.shape[0]
    #print("Trajectory acquired for particle "+str(i))
    T=np.arange(dureetrajectoire)/frequence
    if dureetrajectoire>0:
        return x1-x1[0],y1-y1[0],T,dureetrajectoire
    else:
        return [0],[0],T,dureetrajectoire
def concatenate(ilist=np.arange(Nbobjets)+1):
    X=[0]
    Y=[0]
    N=0
    for i in ilist:
        a=take(i)
        if a[-1]>longueurminimale:
            x1,y1,t,n=a
            X+=list(x1-x1[0]+X[-1])[1:]
            Y+=list(y1-y1[0]+Y[-1])[1:]
            N+=n
    T=np.arange(N)/frequence
    return np.array(X),np.array(Y),T,N
def plot(X,Y,ax,fig):
    t = np.linspace(0,1,X.shape[0])
    dx=X[1:]-X[:-1]
    dy=Y[1:]-Y[:-1]
    dr=(dx**2+dy**2)**0.5
    dr1=dr/dr.max()
    points = np.array([X,Y]).transpose().reshape(-1,1,2)
    segs = np.concatenate([points[:-1],points[1:]],axis=1)
    lc = LineCollection(segs, cmap=plt.get_cmap('jet'))
    lc.set_array(dr)
    plt.gca().add_collection(lc)
    ax.set_xlim(X.min(), X.max())
    ax.set_ylim(Y.min(), Y.max())
    cbar = fig.colorbar(lc,label="Déplacement entre 2 images ($\mathrm{\mu}$m)")
    ax.axis('equal')
    ax.set_xlabel("Déplacement selon X ($\mathrm{\mu}$m)")
    ax.set_ylabel(r"Déplacement selon Y ($\mathrm{\mu}$m)")

def newfig(nrow=1,ncol=1):
    fig,axes=plt.subplots(nrow,ncol)
    fig.set_dpi(192)
    return fig,axes
def autocorr(x,formsd=False):
    fftx = np.fft.fft(x, n=(x.shape[0]*2-1), axis=0)
    cfftx=np.conjugate(fftx)
    ret = np.real(np.fft.ifft(fftx*cfftx, axis=0))
    result = np.fft.fftshift(ret, axes=0)
    result = np.correlate(x, x, mode='full')
    if formsd:
        N=result.size//2+1
        return result[result.size//2:]/(N-np.arange(0,N))
    return result[result.size//2:]
def msd(r):
    N=r.shape[0]
    D=r**2
    D=np.append(D,0)
    S2=autocorr(r,True)
    Q=2*D.sum()
    S1=np.zeros(N)
    for m in range(N):
        Q=Q-D[m-1]-D[N-m]
        S1[m]=Q/(N-m)
    return np.arange(N)/frequence,S1-2*S2
def GaussianHisto(dx,n=30):
    """
    Fonction qui calcule l'histogramme supposé gaussien et le fite
    """
    a,b=np.histogram(dx,n,normed=False)
    ok=a>0
    x,y=((b[1:]+b[:-1])/2)[np.where(ok)],np.log(a[np.where(ok)])
    p = np.poly1d(np.polyfit(x,y,2,w=y))
    if p[2]>0:
        p[2]=-np.inf
    return x,y,p
def sigma(p):
    return (-1/p[2]/2)**0.5
def mean(p):
    return -p[1]/2/p[2]
def delta(x,tau):
    return x[tau:]-x[:-tau]
```

```python
colors=["blue","green","red"]
fig,ax=newfig(1,1)
k=0
for i in [1,100,1000]:
    ax.plot(T[:N//2],delta(X,i)[:N//2],color=colors[k],label=r"$\tau$="+str(i/frequence)+" s",linewidth=1)
    k+=1
ax.set_xlabel(r"Temps ($\mathrm{\mu}$s)")
ax.set_ylabel(r"$\Delta X_{\tau}(t)$ ($\mathrm{\mu}$m)")
handles, labels = ax.get_legend_handles_labels()
#ax.set_xlim(0,5)
ax.legend(handles,labels)
fig.savefig("images/DeltaX(tau).pdf")
```

```python
fig,ax=newfig(1,3)
k=0
def format(num):
    string= "%3.2f"%(num,)
    if string[1:]=="%3.2f"%(0,):
        return "%3.2f"%(0,)
    else:
        return string
for i in [1,100,1000]:
    dx=delta(X,i)
    n,bins,patches=ax[k].hist(dx,20,density=True)
    bin_centres = (bins[:-1] + bins[1:])/2
    def gauss(x, *p):
        A, mu, sigma = p
        return A*np.exp(-(x-mu)**2/(2.*sigma**2))
    coeff, var_matrix = curve_fit(gauss, bin_centres, n,p0= [1., 0., 1.])
    xaxis=np.linspace(bin_centres[0],bin_centres[-1],100)
    mu,sigma=coeff[1:]
    ax[k].plot(xaxis,gauss(xaxis, *coeff),label=r"$\mu$ = "+format(mu)+"\n"+r"$\sigma$ = "+format(sigma))
    ax[k].set_ylim(0,max(n)*1.3)
    mx=np.max(np.abs(bin_centres))
    ax[k].set_xlim(-mx,mx)
    ax[k].set_xlabel(r"$\Delta X$ ($\mathrm{\mu}$m)")
    ax[k].set_title(r"$\tau$ = "+str(i/frequence)+ " (s)")
    handles, labels = ax[k].get_legend_handles_labels()
    ax[k].legend(handles,labels)
    k+=1
fig.tight_layout()
fig.savefig("images/Statistique(tau).pdf")
```

```python
file, Nbobjets=init("results/prof.mat")
frequence=100
longueurminimale=1000
binning=2
agrandissement=10
def takebac(i):
    """La fonction qui tracque la particule i"""
    PixelSize=binning*(6450/agrandissement)*1e-3 # en um
    x1=file['objs_link'][0,:][np.where(file['objs_link'][5,:]==i)]*PixelSize
    y1 = file['objs_link'][1,:][np.where(file['objs_link'][5,:]==i)]*PixelSize
    dureetrajectoire=x1.shape[0]
    print("Trajectory acquired for particle "+str(i))
    T=np.arange(dureetrajectoire)/frequence
    if dureetrajectoire>0:
        return x1-x1[0],y1-y1[0],T,dureetrajectoire
    else:
        return [0],[0],T,dureetrajectoire
```

```python
I=2
MAIN=[1,16,11,5]
fig,ax=newfig(2,2)
def bacplot(X,Y,ax,fig):
    t = np.linspace(0,1,X.shape[0])
    dx=X[1:]-X[:-1]
    dy=Y[1:]-Y[:-1]
    dr=(dx**2+dy**2)**0.5
    points = np.array([X,Y]).transpose().reshape(-1,1,2)
    segs = np.concatenate([points[:-1],points[1:]],axis=1)
    lc = LineCollection(segs, cmap=plt.get_cmap('jet'))
    lc.set_array(dr)
    ax.add_collection(lc)
    ax.set_xlim(X.min(), X.max())
    ax.set_ylim(Y.min(), Y.max())
    ax.axis('equal')
    return lc
for i in range(4):
    X,Y,T,N=takebac(MAIN[i]);
    axes=ax.reshape(-1)[i]
    lc=bacplot(X,Y,axes,fig)
fig.subplots_adjust(wspace=.3)
fig.text(0.62, 0.03, 'Circulaires', ha='center')
fig.text(0.25, 0.03, 'Ballistiques', ha='center')
cbar = fig.colorbar(lc,label="Déplacement entre 2 images ($\mathrm{\mu}$m)",ax=ax.ravel().tolist())
fig.savefig("images/Trajectoire_bacterie.pdf")
```
